```python
from typing import Any, Callable, Dict, List, Literal, Optional, Sequence, Tuple, Union, Annotated
```

#### 1. Built-in Primitive Types
##### Defines standard single-value variables.

```python
age: int = 25
name: str = "Alice"
is_active: bool = True
pi: float = 3.14159
```

#### 2. Modern Built-in Collections (Python 3.9+)
##### Uses standard types instead of importing uppercase names from 'typing'.

```python
numbers: list[int] = [1, 2, 3]
user_scores: dict[str, float] = {"Alice": 9.5, "Bob": 8.0}
coordinates: tuple[int, int] = (10, 20)
unique_ids: set[int] = {101, 102, 103}
```

#### 3. Union Type / Optional Type (Python 3.10+)
##### The '|' operator replaces Union and Optional for multiple allowed types.

```python
def process_data(data: int | str) -> None:
    """Accepts either an integer or a string."""
    print(data)
# 'None' indicates the value can be missing (replaces Optional[str]).
middle_name: str | None = None
```

#### 4. Any Type
##### Disables type checking for a variable; accepts any data type.

```python
dynamic_value: Any = "Could be anything"
dynamic_value = 42
```

#### 5. Type Aliases
##### Creates a reusable custom name for complex type signatures.

```python
Point2D = tuple[float, float]
start_location: Point2D = (0.0, 5.5)
```

#### 6. Callable (Functions)
##### Defines expected input types and the return type for a function argument.

```python
def execute_operation(func: Callable[[int, int], int], x: int, y: int) -> int:
    """Accepts a function that takes two ints and returns an int."""
    return func(x, y)
```

#### 7. Sequence (Duck Typing / Interfaces)
##### Accepts any continuous data structure supporting indexing (list, tuple, str).

```python
def print_first_element(items: Sequence[str]) -> None:
    """Works with lists, tuples, or sequences of strings."""
    if items:
        print(items[0])
```

#### 8. Literal Types (Python 3.8+)
##### Restricts a variable to specific, exact values.

```python
def set_theme(theme: Literal["light", "dark"]) -> None:
    """Only allows 'light' or 'dark' as arguments."""
    print(f"Setting theme to {theme}")
```

#### 9. Classes as Types
##### User-defined classes can be used directly as type hints.

```python
class User:

    def __init__(self, username: str):
        self.username = username

def greet_user(user: User) -> str:
    """Accepts an instance of the User class."""
    return f"Hello, {user.username}"
```

#### 9. Annotating with Type Hints
##### User-defined classes can be used directly as type hints.

```python
def say_hello(name: Annotated[str, "this is just metadata"]) -> str:
    return f"Hello {name}"
```
