Modern versions of Python have a very intuitive way to define asynchronous code.
This makes it look just like normal "sequential" code and do the "awaiting" for you at the right moments.
- When there is an operation that will require waiting before giving the results use `await`. It tells Python that it has to wait for the operation to finish,Python will know that it can go and do something else in the meanwhile 
- For `await` to work, it has to be inside a function that supports this asynchronicity. To do that, you just declare it with `async def` instead of just `def`.
- When you want to call an async def function, you have to "await" it.

```python
import asyncio
import time

async def worker(name, delay):
    print(f"{name} started (sleep {delay}s)")
    await asyncio.sleep(delay)
    print(f"{name} finished")
    return name

async def sequential():
    start = time.perf_counter()
    result_a = await worker("A", 1)
    result_b = await worker("B", 1)
    print("Sequential duration:", round(time.perf_counter() - start, 3), "s")
    print("Sequential results:", result_a, result_b)

async def concurrent():
    start = time.perf_counter()
    results = await asyncio.gather(worker("A", 1), worker("B", 1))
    print("Concurrent results:", results)
    print("Concurrent duration:", round(time.perf_counter() - start, 3), "s")

async def main():
    print("Running sequential example")
    await sequential()
    print("\nRunning concurrent example")
    await concurrent()

# In a Jupyter notebook you can use top-level await:
await main()
```
